```python
ys_theta = pickle.load(open('ys_over_models_theta.p', 'rb'))
ys_alpha = pickle.load(open('../ys_over_models_alpha.p', 'rb'))

ys_alpha = dict(zip([x[0] for x in ys_alpha], [np.array(x[1])[relevant_index] for x in ys_alpha])) 
ys_theta = dict(zip([x[0] for x in ys_theta], [x[1] for x in ys_theta])) 
```

```python
for i in range(41):
    try:
        pickle.dump([ys_alpha[i], ys_theta[i], big_df['class'].map(colormap).values], open(f'/home/pbcalder/Documents/analyse_bayesian/PERFORMANCE_PLOTTER/heatmap_dat/cnix_{i}.p', "wb"))
    except:
        continue
```

```python
for i in {0, 2, 5, 7, 10, 11, 16, 18, 19, 20, 24, 27, 28, 29, 30, 32, 35, 36}:
    try:
        ax = sns.jointplot(
            x = (ys_alpha[i]), 
            y = (ys_theta[i]),
            hue = big_df['class'].map(colormap).values,
            kind="kde",
            alpha=0.9,
            palette=['gray', 'pink','red'],
            hue_order=['neutral', 'clickbait', 'inflammatory'],
            marginal_kws={'common_norm':False},
            joint_kws={'levels':10}
        )
        ax.set_axis_labels(xlabel=f"log cascade size\n\n{model_index_to_name[i]}", ylabel=f'log halflife')
        plt.savefig(f"jointgrid/{i}_{model_index_to_name[i]}.png", bbox_inches='tight')
        plt.close()
    except:
        continue
```

```python
for i in {0, 2, 5, 7, 10, 11, 16, 18, 19, 20, 24, 27, 28, 29, 30, 32, 35, 36}:
    try:
        ax = sns.jointplot(
            x = np.exp(ys_alpha[i]), 
            y = np.exp(ys_theta[i]),
            hue = big_df['class'].map(colormap).values,
            kind="kde",
            alpha=0.9,
            palette=['gray', 'pink','red'],
            hue_order=['neutral', 'clickbait', 'inflammatory'],
            marginal_kws={'common_norm':False}
        )
        ax.set_axis_labels(xlabel=f"cascade size\n\n{model_index_to_name[i]}", ylabel=f'halflife')
        plt.savefig(f"jointgrid_exp/{i}_{model_index_to_name[i]}.png", bbox_inches='tight')
        plt.close()
    except:
        continue
```

```python
a_s = []
t_s = []
col_headers = []
for i in {0, 2, 5, 7, 10, 11, 16, 18, 19, 20, 24, 27, 28, 29, 30, 32, 35, 36}:
    a = pd.DataFrame(ys_alpha[i])
    t = pd.DataFrame(ys_theta[i])

    a_s.append(a)
    t_s.append(t)
    col_headers.append(model_index_to_name[i])

as_df = pd.concat(a_s, axis=1)
ts_df = pd.concat(t_s, axis=1)
as_df.columns = col_headers
ts_df.columns = col_headers
as_df.to_csv("pred_alpha_fakenix.csv", index=False)
ts_df.to_csv("pred_theta_fakenix.csv", index=False)
```
